```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%pylab 
%matplotlib inline
import statsmodels.api as sm
import seaborn as sns
sns.set_context("talk")
from IPython.display import HTML
import folium
```

```python

coffee = pd.read_csv("coffee.csv")
Productivity=pd.read_csv("Productivity.csv", skiprows=4)

countries = ['Colombia' ,'United States' , 'Germany', 'Japan', 'France', 'Italy']


for i in range(len(Productivity['Country Name'])):
    country = Productivity["Country Name"][i]
    if country not in countries:
        Productivity = Productivity.drop([i])

Productivity.reset_index(drop = True, inplace = True)



#coffee = pd.melt(coffee, id_vars=coffee.columns[0:6], var_name ="Year", value_name = "Value")
for i in range(len(coffee['Region'])):
    country = coffee["Region"][i]
    if country not in countries:
        coffee = coffee.drop([i])


coffee.reset_index(drop = True, inplace = True)
#coffee_volume['Value'] = coffee_volume['Value'].str.replace(',','')
#coffee_volume = coffee_volume.astype({"Value":"float","Year":"int"})
#coffee_volume= coffee_volume.rename(columns={'Value': 'Volume'})
coffee_volume=coffee.drop(columns=["Market","Name", "Unit","Source"])
```

```python
hours_worked = pd.read_csv('ANHRS_08112022231653052.csv')
hours_worked = hours_worked[["Country", "Time", "Value"]]
countries = [ 'Colombia' , 'United States' , 'Germany', 'Japan', 'France', 'Italy']

for i in range(len(hours_worked['Country'])):
    country = hours_worked["Country"][i]
    if country not in countries:
        hours_worked = hours_worked.drop([i])

hours_worked=hours_worked.reset_index(drop=True)

years=[2013,2014,2015,2016,2017,2018,2019,2020]
for i in range(len(hours_worked['Time'])):
    year = hours_worked["Time"][i]
    if year not in years:
        hours_worked = hours_worked.drop([i])

hours_worked=hours_worked.reset_index(drop=True)
hours_worked= hours_worked.rename(columns={'Value': 'Hours Worked'})


energy_drinks=pd.read_csv("energy_drinks.csv")

coffee_volume=coffee_volume[['Region','Chart','2013', '2014', '2015', '2016', '2017', '2018','2019', '2020']]
energy_drinks=energy_drinks[['Region','Chart','2013', '2014', '2015', '2016', '2017', '2018','2019', '2020']]
Productivity=Productivity[['Country Name','2013', '2014', '2015', '2016', '2017', '2018','2019', '2020']]

countries = ['Colombia' , 'United States' , 'Germany', 'Japan', 'France', 'Italy']

energy_drinks=energy_drinks.reset_index(drop=True)

for i in range(len(energy_drinks['Region'])):
    country = energy_drinks["Region"][i]
    if country not in countries:
        energy_drinks = energy_drinks.drop([i])

```

```python
energy_drinks=energy_drinks.T
new_header = energy_drinks.iloc[0] #grab the first row for the header
energy_drinks = energy_drinks[1:] #take the data less the header row
energy_drinks.columns = new_header
energy_drinks=energy_drinks.reset_index()
energy_drinks = energy_drinks.rename(columns={'index': 'Time'})
energy_drinks = energy_drinks[1:]




Prod=Productivity.T
new_header = Prod.iloc[0] #grab the first row for the header
Prod = Prod[1:] #take the data less the header row
Prod.columns = new_header
Prod=Prod.reset_index()

coffee_volume=coffee_volume.T
new_header = coffee_volume.iloc[0] #grab the first row for the header
coffee_volume = coffee_volume[1:] #take the data less the header row
coffee_volume.columns = new_header
coffee_volume=coffee_volume.reset_index()
coffee_volume = coffee_volume.rename(columns={'index': 'Time'})
coffee_volume = coffee_volume[1:]
#coffee_volume
energy_drinks
```

```python

#France 
data_France_HW=hours_worked["Country"]=='France'
France_HW=hours_worked[data_France_HW]
France_HW=France_HW.drop([8,9,10,11,12,13,14,15])


data_France_ED=energy_drinks["France"]
data_France_Prod=Prod["France"]
data_France_Coffee=coffee_volume["France"]
data_France_Coffee

France = pd.concat([France_HW, data_France_ED], axis=1, join='inner')
France = pd.concat([France, data_France_Prod], axis=1, join='inner')
France = pd.concat([France, data_France_Coffee], axis=1, join='inner')
France = pd.DataFrame(data=France)
France.columns = ['Country', 'Time', 'Hours Worked', 'Energy Drink Revenue', 'Energy Drink Volume', 'GDP', "Coffee Revenue", "Coffee Volume"]
%store France
France
```
