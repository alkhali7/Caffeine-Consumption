To find how coffee consumption affected hours worked, we first had to clean the datasets. The hours worked data was missing 6 countries, so we decided to analyze only France, the United States, Germany, Italy, Japan, and Colombia. Additionally, we spent a lot of time unifying the datasets. For example, the timeframes of the coffee consumption datasets and the hours worked datasets were different so we had to find a common timeframe.  Another issue was the coffee consumption data had different years spread out as columns, while hours worked had a single years column. We used pandas to clean and unify the data.

### Hours Worked Cleaning 

```python
hours_worked = pd.read_csv('ANHRS_08112022231653052.csv')

# only look at total employment data 
hours_worked = hours_worked[hours_worked["Employment status"] == "Total employment"]

# have to reset the index after dropping things 
hours_worked.reset_index(drop = True, inplace = True)

# reducing dataframe 
hours_worked = hours_worked[["Country", "Time", "Value"]]

countries = ['Brazil' , 'Vietnam', 'Colombia' , 'Indonesia' , 'Ethiopia', 'Honduras', 
             'United States' , 'Germany', 'Japan', 'France', 'Italy', 'Russia']

for i in range(len(hours_worked['Country'])):
    country = hours_worked["Country"][i]
    if country not in countries:
        hours_worked = hours_worked.drop([i])
        
hours_worked.reset_index(drop = True, inplace = True)
```

### Coffee Volume Cleaning

```python
coffee = pd.read_csv("coffee.csv")

# change coffee data to a long format 
coffee = pd.melt(coffee, id_vars=coffee.columns[0:6], var_name ="Year", value_name = "Value")

# only use the countries that are in hours_worked 
countries = hours_worked["Country"].unique()
for i in range(len(coffee['Region'])):
    country = coffee["Region"][i]
    if country not in countries:
        coffee = coffee.drop([i])
        
# only look at volume data 
coffee_volume = coffee[coffee["Chart"] == "Volume"]

# reset the indices 
coffee_volume.reset_index(drop = True, inplace = True)

# get rid of commas in value 
coffee_volume['Value'] = coffee_volume['Value'].str.replace(',','')

# change the value and year columns to numerics 
coffee_volume = coffee_volume.astype({"Value":"float","Year":"int"})
```

### Unify coffee_volume and hours_worked dataframes

```python
hours_worked = hours_worked.sort_values("Time")

# coffee volume goes from 2012 to 2025 (predicted)
# hours_worked goes from 2010 to 2021
# time interval should be 2012 to 2020 
years = np.arange(2012, 2022)

# filter coffee volume years 
for i in range(len(coffee_volume['Year'])):
    year = coffee_volume["Year"][i]
    if year not in years:
        coffee_volume = coffee_volume.drop([i])
        
# filter hours_worked years 
for i in range(len(hours_worked['Time'])):
    year = hours_worked["Time"][i]
    if year not in years:
        hours_worked =hours_worked.drop([i])
        
hours_worked.reset_index(drop = True, inplace = True)

#hours_worked.head()
```

```python
# since colombia data in hours worked only goes from 2015 to 2020, filter coffee_volume accordingly
for i in range(len(coffee_volume['Region'])):
    year = coffee_volume["Year"][i]
    country = coffee_volume["Region"][i]
    if year not in np.arange(2015, 2021):
        if country == "Colombia":
            coffee_volume = coffee_volume.drop([i])
            
hours_worked.reset_index(drop = True, inplace = True)
#coffee_volume.head()
```

## Question 4: How does the volume of energy drinks consume affect productivity(GDP)? 

To figure out how energy drink consumption affects productivity we cleaned up our data.  First, we read in our Productivity dataset using pandas read csv.  After we did that we dropped all of the columns that did not factor in nor contribute to our question.  These were "Country Code", "Indicator Name", "Indicator Code", "Unnamed: 66", and other years that we did not want to measure.  After doing this we read in the Energy Drinks dataset using pandas.  Then we cleaned out that data frame by dropping the commas of the string value from every year.

Once the data was cleaned we created a dictionary for each of the countries.  Then by using a loop we appended the GDP and energy drink data to the country dictionary.  After the dictionary was created we created a function that displays the correlation.  The function took in the volume of energy drinks that each country consumed and the GDP for each country then put them into separate data frames.  Then it created a model for the country and ran an OLS regression for the model using the class from statsmodel.  The function then returned the results for the fit.  

Once we had our correlation function we created regression models for each country.  Which we used to find the R square values, F statistics, and significance. With these models we used the function sm.graphics.plot_regress_exog to plot our linear regressions. 

#### Reading in and Cleaning Data

```python
# First the productivity file
countries = ["Brazil" , "Vietnam", "Colombia" , "Indonesia" , "Ethiopia", "Honduras", "United States" , "Germany", "Japan", "France", "Italy", "Russia" ]
productivity_gdp = pd.read_csv("Productivity.csv", skiprows=4)
productivity_gdp.head()

for i in range (len(productivity_gdp["Country Name"])) :
    country = productivity_gdp["Country Name"][i]
    if country not in countries:
        productivity_gdp = productivity_gdp.drop([i])

productivity_gdp=productivity_gdp.drop(columns=["Country Code","Indicator Name", "Indicator Code", "Unnamed: 66"], axis=1)
productivity_gdp=productivity_gdp.drop(productivity_gdp.iloc[:, 1:50], axis=1)
productivity_df = pd.DataFrame(productivity_gdp)
#productivity_df
```

```python
#now the energy drink dataset
energy_drink_data  = pd.DataFrame(pd.read_csv("energy_drinks.csv"))

# First make
energy_drink_data['2013'] = energy_drink_data['2013'].str.replace(',','')
energy_drink_data['2014'] = energy_drink_data['2014'].str.replace(',','')
energy_drink_data['2015'] = energy_drink_data['2015'].str.replace(',','')
energy_drink_data['2016'] = energy_drink_data['2016'].str.replace(',','')
energy_drink_data['2017'] = energy_drink_data['2017'].str.replace(',','')
energy_drink_data['2018'] = energy_drink_data['2018'].str.replace(',','')
energy_drink_data['2019'] = energy_drink_data['2019'].str.replace(',','')
energy_drink_data['2020'] = energy_drink_data['2020'].str.replace(',','')
energy_drink_data['2021'] = energy_drink_data['2021'].str.replace(',','')

#energy_drink_data
```

```python
# Next, I created a dictionary to hold countries with the volume consumed for each year from 2013-2021 and the GDP from 2013-2021
country_dict = {}
for country in productivity_gdp["Country Name"]:
    country_dict[country] = {"GDP":[], "Volume":[], "Change in GDP":[]}

    # working with productivity_gdp and energy_drink_data   
    #https://stackoverflow.com/questions/51500889/python-valueerror-can-only-convert-an-array-of-size-1-to-a-python-scalar-whe
    country_dict[country]["GDP"].append(float(productivity_gdp[productivity_gdp["Country Name"]==country]['2012'].tolist()[0]))
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2013'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2014'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2015'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2016'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2017'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2018'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2019'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2020'].tolist()[0])
    country_dict[country]["GDP"].append(productivity_gdp[productivity_gdp["Country Name"]==country]['2021'].tolist()[0])
    
    #print(energy_drink_data[energy_drink_data["Region"]==country and energy_drink_data["Chart"]=="Volume"]['2013'])
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2013'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2014'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2015'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2016'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2017'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2018'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2019'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2020'].tolist()[1]) * 10e6)
    country_dict[country]["Volume"].append(float(energy_drink_data[energy_drink_data["Region"]==country]['2021'].tolist()[1]) * 10e6)
    
    i = 0
    j = 1
    while j < len(country_dict[country]["GDP"]):
        country_dict[country]["Change in GDP"].append(country_dict[country]["GDP"][j] - country_dict[country]["GDP"][i])
        i+=1
        j+=1
    country_dict[country]["GDP"].pop(0) #we don't want to include 2012 in final data
    
#country_dict
```

#### Regression 

```python
# Create a function to perform the regression and return the result for each country by using the dictionary we created above
def normal_ols_correlation(c_dict, country):
    temp_dict = {"Volume": c_dict[country]["Volume"], "GDP": c_dict[country]["GDP"]}
    country_df = pd.DataFrame(temp_dict)
    country_df.astype("float64").dtypes
    
    volume_df = pd.DataFrame(c_dict[country]["Volume"])
    volume_df = volume_df.rename({0: "Volume"}, axis = 1)
    type_gdp_df = pd.DataFrame(c_dict[country]["GDP"])
    type_gdp_df = type_gdp_df.rename({0: "GDP"}, axis = 1)
    model = sm.OLS(volume_df, type_gdp_df)
    results = model.fit()
    return results, volume_df
```
